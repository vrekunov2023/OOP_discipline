## Порівняння простих даних з обєктами

```python
class MyAnimals:
    def __init__(self, animals, name, age, mass = None) -> None:
        print(f"В мене є тваринка {animals} яку звати {name} які {age} років")
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass==None else mass


pet1 = MyAnimals("Рибки", "Неони", 1)
pet2 = MyAnimals("Кіт", "Мурка", 5, 10)
pet3 = MyAnimals(animals="Собака", name="Рекс", age=2)

# вивести об'єкти та їх атрибути
print(f"pet1: {pet1.animals}, {pet1.name}, {pet1.age} років, маса {pet1.mass} кг")
print(f"pet2: {pet2.animals}, {pet2.name}, {pet2.age} років, маса {pet2.mass} кг")
print(f"pet3: {pet3.animals}, {pet3.name}, {pet3.age} років, маса {pet3.mass} кг")

print(pet2.name)
pet2.name = "Машка"
print(f"Я переіменував свою тваринку на {pet2.name}, вона важить {pet2.mass} кг")

print(f"Моя тваринка {pet1.name}, вона важить {pet1.mass} кг")
```

## Методи класу

```python
class MyAnimals:
    def __init__(self, animals, name, age, mass = None) -> None:
        print(f"В мене є тваринка {animals} яку звати {name} якій {age} років")
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass==None else mass
    
    def speak(self):
        if self.animals == "Кіт":
            return f"Моя Тваринка {self.animals} говорить Мяууу"
        elif self.animals == "Собака":
            return f"Моя Тваринка {self.animals} говорить Гавв"
        else:
            return f"Моя Тваринка {self.animals} не розмовляє"
    
    def add_one_year(self):
        self.age += 1
        return f"Моя Тваринка {self.animals} прожила зі мною один рік"

pet1 = MyAnimals("Рибки", "Неони", 1)
pet2 = MyAnimals("Кіт", "Мурка", 5, 10)
pet3 = MyAnimals(animals="Собака", name="Рекс", age=2)

print(f"Тваринка {pet1.animals}:", pet1.speak())

print(f"Тваринка {pet3.animals}:", pet3.speak())


print(f"{pet1.animals} живе зі мною {pet1.age} років")
print(pet1.add_one_year())
print(f"{pet1.animals} живе зі мною {pet1.age} років")

print(pet1.age)
```

## Змінні класу, глобальні змінні над всіма обєктами

```python
class MyAnimals:
    total_animals = 0
    total_animal_types = set()

    def __init__(self, animals, name, age, mass = None) -> None:
        print(f"В мене є тваринка {animals} яку звати {name} які {age} років")
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass==None else mass
        MyAnimals.total_animal_types.add(animals)
        MyAnimals.total_animals += 1

# початковий стан (до створення нових об'єктів)
print(f"В мене зараз є тваринки {list(MyAnimals.total_animal_types)}, їх кількість {MyAnimals.total_animals}")

# створити декілька об'єктів в циклі
pets_data = [
    ("Рибки", "Неони", 1),
    ("Рибки", "Гупії", 1),
    ("Рибки", "Соміки", 1),
    ("Кіт", "Мурка", 5, 10),
    ("Собака", "Рекс", 2)
]

created_pets = []
for item in pets_data:
    if len(item) == 3:
        animals, name, age = item
        mass = None
    else:
        animals, name, age, mass = item
    created_pets.append(MyAnimals(animals, name, age, mass))

# вивести глобальні (класові) змінні після створення об'єктів
print(f"Після циклу: тваринки {list(MyAnimals.total_animal_types)}, їх кількість {MyAnimals.total_animals}")

# приклад доступу до класової змінної через об'єкт та демонстрація shadowing
print(f"Доступ через об'єкт: {created_pets[0].total_animal_types}")
created_pets[0].total_animal_types = {"Папуга"}  # тіньове присвоєння на рівні об'єкта
print(f"Після тіньового присвоєння: об'єкт {created_pets[0].total_animal_types}, клас {MyAnimals.total_animal_types}")
```

## Статичні методи та змінна self

```python
class MyAnimals:
        def __init__(self, animals, name, age) -> None:
            self.animals = animals
            self.name = name
            self.age = age
        
        def add_one_year(self):
            self.age += 1
            return f"Моя Тваринка {self.animals} прожила за мною один рік"

        @staticmethod
        def call_pet():
            return "Тваринка біжить до нас"


pet1 = MyAnimals("Рибки", "Гупії", 1)
pet2 = MyAnimals("Кіт", "Мурка", 5)

# виклик статичного методу для кожного об'єкта
print(f"Статичний виклик для pet1: {pet1.call_pet()}")
print(f"Статичний виклик для pet2: {pet2.call_pet()}")

print(f"Ми кличемо нашу тваринку за допомогою статичного методу з обєкту |{pet1.call_pet()}| та з класу |{MyAnimals.call_pet()}|")

print(f"Викликаю метод з обєкту |{pet1.add_one_year()}| та з класу |{MyAnimals.add_one_year(pet1)}|")
```

## Робота з properties

```python
from datetime import datetime
class MyAnimals:
    max_age = 20
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
    
    @property
    def live_untill(self):
        return datetime.today().year - self.age + self.max_age
    
    @property
    def years_left_to_live(self):
        return self.max_age - self.age
    
    def add_one_year(self):
        self.age += 1
        return f"Моя Тваринка {self.animals} прожила зі мною один рік"


pet1 = MyAnimals("Кіт", "Мурка", 5)
print(f"Мій {pet1.animals} будже зі мною до {pet1.live_untill} року")
print(f"У моєї {pet1.animals} залишилось {pet1.years_left_to_live} років життя")
pet1.add_one_year()
print(f"Мій {pet1.animals} будже зі мною до {pet1.live_untill} року")
print(f"У моєї {pet1.animals} залишилось {pet1.years_left_to_live} років життя")
```

## Методи Класу

```python
class MyAnimals:
    total_csv = int(0)
    total_objects = int(0)
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
        MyAnimals.total_objects += 1

    @classmethod
    def from_csv(cls, csv_data:str):
        cls.total_csv += 1
        animal, name, age = csv_data.split(",")
        return cls(animal, name, age)
    
    @classmethod
    def is_name_capitalized(cls, name: str):
        return name[0].isupper()

a1 = MyAnimals.from_csv("Кіт,Мурка,5")
a2 = MyAnimals("Собака", "Рекс", 2)

print(f"Мій {a1.animals}, звати {a1.name}, має {a1.age} років")
print(f"Мій {a2.animals}, звати {a2.name}, має {a2.age} років")
print(f"Загальна кількість обєктів {MyAnimals.total_objects} а створених з csv формату {MyAnimals.total_csv}")

# перевірка namen на велику букву
print(f"Ім'я '{a1.name}' починається з великої букви: {MyAnimals.is_name_capitalized(a1.name)}")
print(f"Ім'я '{a2.name}' починається з великої букви: {MyAnimals.is_name_capitalized(a2.name)}")
```

## Приватні атрибути та методи

```python
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
        self._secret_phrase = "Це приватний атрибут, (як я буду кликати свою тваринку)"
        self.__secret_strong = "Цей атрибут точно не можна викликати ззовні!!"
    
    @property
    def secret_phrase(self):
        return f"Ми доступаємось до приватного атрибуту з проперті: {self._secret_phrase}"
    
    @property
    def secret_strong(self):
        return f"Ми доступаємось до приватного атрибуту з проперті: {self.__secret_strong}"

    def _secret_methid(self):
        return "Приватний метод, краще його не рухати, він щось робить потрібне всередині класу"

    def __do_not_touch_it(self):
        return "Це неможливо викликати ззовні"
    


a1 = MyAnimals("Собака", "Рекс", 2)

print(a1._secret_phrase)
a1._secret_phrase = "Я змінив приватний атрибут"  ## Такого робити не треба!
print(a1._secret_phrase)

## a1.__secret_strong ## Це видасть помилку!
print(a1._MyAnimals__secret_strong) ### Так також робити не можна!

### Виводимо приватний атрибут через проперті
print(a1.secret_phrase)
#a1.secret_phrase = "З проперті так зробити не можна! це видасть помилку"

print(a1.secret_strong)

print(a1._secret_methid())

#a1.__do_not_touch_it() ##Це виклик видасть помилку!
```

## Магічні методи

```python
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    def __repr__(self) -> str:
        return f'MyAnimals("{self.animals}", "{self.name}", {self.age})'

    def __str__(self) -> str:
        return f"Це є обєкт {self.__repr__()}"

a1 = MyAnimals("Собака", "Рекс", 2)
print(a1)

print(repr(a1))

str(a1)
```

## Арифметичні методи

```python
a = 2
b = 5
c = "C"

print(a+b, a - int(1),  a * b, str(a) + c)

print(a.__add__(b), a.__sub__(int(1)), a.__mul__(b))

class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    def __sub__(self, obj):
        if isinstance(obj, MyAnimals):
            return f"Мій {self.name} покусав {obj.name}"
        elif isinstance(obj, int):
            self.age += obj
            return f"Мій {self.name} постарів на {obj} років, тепер йому {self.age}"
        return NotImplementedError
    
    def __gt__(self, obj):
        if isinstance(obj, MyAnimals):
            if self.age > obj.age:
                return f"Старшим є {self.name}"
            return f"{self.name} не є старшим за {obj.name}"
        return NotImplementedError

    def __len__(self) -> int:
        print("Ми зайшли в метон len, та знаходимо довжину Імені тваринки")
        return len(self.name)
            
        

a = MyAnimals("Собака", "Рекс", 2)
b = MyAnimals("Кіт", "Мурзик", 4)
print(a - b)
print(a - 1)
print(a - float(1))

print(a > b)

print(b > a)

len(a)
```
